# やじるし

```python
# 矢印の始点と終点
start = [0, 0]
end = [1, 1]

# 背景をなくした矢印を描画
plt.figure(figsize=(5, 5))
plt.quiver(*start, *(np.array(end) - np.array(start)), angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')  # 背景をなくす
plt.show()
```

```python
# 各軸の始点と方向ベクトル
origins = np.array([[0, 0, 0]] * 3)  # 全ての矢印の始点
vectors = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])  # X, Y, Z軸方向のベクトル
colors = ['red', 'green', 'blue']  # 各軸の色

# 3Dプロットの設定
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# 各軸をプロット
for origin, vector, color in zip(origins, vectors, colors):
    ax.quiver(*origin, *vector, color=color, linewidth=2)

# プロット範囲と設定
ax.set_xlim([-0.5, 1.5])
ax.set_ylim([-0.5, 1.5])
ax.set_zlim([-0.5, 1.5])
ax.set_axis_off()  # 背景を非表示

# 表示
plt.show()
```

# Higgsポテンシャル

```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ設定
mu = 1.0      # 質量項（スケール）
lambda_ = 0.5 # 相互作用強度

# スカラー場の範囲
phi = np.linspace(-2, 2, 500)

# ヒッグスポテンシャルの計算
V = -mu**2 * phi**2 + lambda_ * phi**4

# 図の描画
plt.figure(figsize=(8, 6))
plt.plot(phi, V, label=r'$V(\phi) = -\mu^2 \phi^2 + \lambda \phi^4$', color="blue")

# タイトルやラベル
plt.title("Higgs Potential", fontsize=16)
plt.xlabel(r'$\phi$', fontsize=14)
plt.ylabel(r'$V(\phi)$', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # x軸
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # y軸
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# グラフの表示
plt.show()
```

```python
# パラメータ設定
mu = 1.0      # 質量項（スケール）
lambda_ = 0.5 # 相互作用強度
# スカラー場の範囲
phi = np.linspace(-2, 2, 500)
# ヒッグスポテンシャルの計算
V = -mu**2 * phi**2 + 0.8*lambda_ * phi**4

ellipse_x = 1.13*np.cos(np.linspace(0, 2 * np.pi, 500))  # 楕円のx座標
ellipse_y = -0.63 + 0.3*np.sin(np.linspace(0, 2 * np.pi, 500))  # 楕円のy座標

ellipse_x2 = 2*np.cos(np.linspace(0, 2 * np.pi, 500))  # 楕円のx座標
ellipse_y2 = 2.4 + 0.3*np.sin(np.linspace(0, 2 * np.pi, 500))  # 楕円のy座標

fig, ax = plt.subplots(figsize=(8, 6))
plt.plot(phi, V, label=r'$V(\phi) = -\mu^2 \phi^2 + \lambda \phi^4$', color="black", linewidth=3)

axiswidth = 0.0035

# 軸
ax.quiver(-2.2, 0, 4.4, 0, angles='xy', scale_units='xy', scale=1, color='black', width=axiswidth)  # Re(\phi)
# ax.text(1.8, -0.2, "$\\mathrm{Re}(\\phi)$", fontsize=14, va="center")

ax.quiver(-0.45, -0.45, 1.5, 1.5, angles='xy', scale_units='xy', scale=1, color='black', width=axiswidth)  # Im(\phi)
# ax.text(0.6, 0.9, "$\\mathrm{Im}(\\phi)$", fontsize=14, ha="center")

ax.quiver(0, -0.5, 0, 3.7, angles='xy', scale_units='xy', scale=1, color='black', width=axiswidth)  # V(\phi)
# ax.text(-0.3, 2.9, "$V(\\phi)$", fontsize=14, ha="center")

# 楕円 (破線)
ax.plot(ellipse_x, ellipse_y, linestyle="--", color="gray")
ax.plot(ellipse_x2, ellipse_y2, linestyle="--", color="gray")

# 真空期待値の点とラベルを追加
ax.plot([1.13], [-0.54], 'bo', markersize=10)  # 青い点
ax.text(1.13, 0.1, "真空期待値", fontsize=12, color="blue", ha="center")

# 両矢印を追加
# ax.annotate("", xy=(0.85, -0.45), xytext=(1.3, -0.45), arrowprops=dict(arrowstyle="<->", color="blue", lw=2))

# 下に凸の矢印を追加
# arrow = FancyArrow(0.85, -0.5, 0.45, 0.0, width=0.05, head_width=0.1, head_length=0.1, color="blue")
# ax.add_patch(arrow)
# arrow = FancyArrow(1.3, -0.5, -0.45, 0.0, width=0.05, head_width=0.1, head_length=0.1, color="blue")
# ax.add_patch(arrow)

# 青い点と文字をつなぐ波線
wave_x = np.linspace(1.13, 1.13, 100)
wave_y = np.linspace(-0.58, 0.06, 100)
# wave_y += 0.05 * np.sin(10 * np.pi * wave_y)  # 波線の調整

ax.plot(wave_x, wave_y, linestyle="--", color="blue", linewidth=1.5)


# 軸範囲と設定
ax.set_xlim(-3, 3)
ax.set_ylim(-1, 4)
ax.set_aspect('equal', adjustable='box')
ax.axis('off')  # 軸を非表示

plt.show()
```
